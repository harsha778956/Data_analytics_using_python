```python
import pandas as pd

data=pd.read_csv("ecommerce_dataset_updated.csv")
data

c_data=data.loc[:,["Product_ID","Category"]]

sales_per_category=c_data.groupby("Category").count()
sales_per_category

import matplotlib.pyplot as plt
import seaborn as sns
print(sales_per_category)

sns.set(style="whitegrid")
# sns.barplot(data=sales_per_category,x="Category")
ax=sns.countplot(data=data,x="Category")
plt.title("Bar graph indicating no of sales of each category")
plt.xticks(rotation=30)
plt.show()
```

```python
sns.set(style="whitegrid")
# sns.barplot(data=sales_per_category,x="Category")
sns.countplot(data=data,x="Payment_Method")
plt.title("Bar graph indicating no of mode of payment")
plt.xticks(rotation=30)
plt.show()
```

```python
avg_price=data.loc[:,['Category',"Final_Price(Rs.)"]].groupby("Category").mean()
avg_price
```

```python
sns.barplot(data=avg_price,x=avg_price.index,y="Final_Price(Rs.)")
plt.title("Bar graph indicating the average price in each category")
plt.xticks(rotation=30)
```

# Based on the data visualizations from your e-commerce dataset, we can draw several inferences:

1. **Sales Distribution by Category**: The bar graph indicates that the "Home & Kitchen" category has the highest number of sales, followed closely by "Books" and "Beauty" categories. This suggests a strong consumer preference for these categories, likely due to their essential nature and frequent usage.

2. **Payment Methods**: The count plot for payment methods shows a diverse range of preferences among consumers. UPI and Net Banking are the most popular payment methods, reflecting the growing trend of digital payments in India. Cash on Delivery also remains a significant mode of payment, highlighting the trust issues some consumers still have with online transactions.

3. **Discount Impact**: Discounts play a crucial role in influencing purchasing decisions. The dataset shows various discount percentages across different categories, with higher discounts likely driving more sales. This is evident from the significant sales in categories like "Clothing" and "Electronics," where discounts are more common.

4. **Average Price Analysis**: The bar graph indicating the average price in each category shows that "Clothing" has the highest average final price, followed by "Books" and "Sports." This could be due to the higher value of individual items in these categories.

5. **Visualizations**: The bar graphs and count plots provide a clear visual representation of the data. The bar graph indicating the number of sales per category helps in understanding the popularity of each category. Similarly, the count plot for payment methods gives insights into consumer payment preferences. The bar graph showing the average price in each category helps in identifying the pricing trends.

These insights can help in making informed decisions regarding inventory management, marketing strategies, and pricing policies.
